```python
import matplotlib.pyplot as plt

print("Hello World")
```

```python
x = [15, 18, 24, 3] # Make a list for X
y = [1, 2, 3, 4] # Make a list for Y

plt.xlabel("Jokes") # Makes X Label 'jokes'
plt.ylabel("Memes") # Makes Y label 'memes'
plt.title("Hot Potato") # Makes plot title 'Hot Potato'

plt.plot(x,y, linestyle = 'dashed') # Plots the X and Y, and makes the style of the line dashed.
```

```python
x = [30,60,180,150] # Makes a list for X
mylabels = ["Apples", "Oranges", "Blackberries", "Blueberries"] # Makes a list for labels of X

plt.pie(x, labels = mylabels, autopct='%1.1f%%') # plots X as a pie chart, makes the label the myLabels (variable), and shows percentage of pie.
```

```python
x = [10,15,20,25,30] # Makes a list for X
y = [20,30,40,50,60] # Makes a list for Y

plt.xlabel("Sadness")
plt.ylabel("Happiness")
plt.bar(x,y) # plots X and Y as a bar graph
```

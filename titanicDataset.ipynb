```
1. What is the percentage of people who had survived?
```

````python
import pandas as pd
import matplotlib.pyplot as plt

died = df[df['Survived'] == 0] # Implements a new dataset for people who died, as 0.
survived = df[df['Survived'] == 1] # Implements a new dataset for people who died, as 0.

died_count = died.shape[0] # Counts rows in died dataset.
survived_count = survived.shape[0] # Counts rows in survived dataset.

labels = ['Died', 'Survived'] # Label for chart
sizes = [died_count, survived_count] # Num of passengers for category.

plt.pie(sizes, labels=labels, autopct='%1.1f%%') # Pie chart
plt.title('Distribution of Passengers') # Pie chart title
plt.show()
````

```
2. What is the average fare of everyone who paid and died, versus the average fare of everyone who paid and was alive, on the titanic dataset?
```


````python
import pandas as pd
import matplotlib.pyplot as plt


# Read the Titanic dataset
df = pd.read_csv("titanic.csv")


# Calculate the average fare for passengers who died

died = df[df['Survived'] == 0]['Fare'].mean() # Creates variable to filter survived dataset of 0, and gets the mean of the fare for the dead
survived = df[df['Survived'] == 1]['Fare'].mean() # Creates variable to filter survived dataset of 1, and gets the mean of the fare for the survived
died = round(died, 2) # Rounds died 2 decimal places
survived = round(survived, 2) # Rounds survived 2 decimal places

print('The mean of the people who paid and died was', died)
print('The mean of the people who paid and survived was', survived)


label = 'Died', "Survived"
ylist = [died, survived]
colors = ['black', 'green']
plt.title('Averages of people who paid')
plt.bar(label, ylist, color = colors)
plt.show()
````

```
3. What is the percentage of people 21 and under who survived, as opposed to the people over 21?
```

````python
import pandas as pd
import matplotlib.pyplot as plt

young = df[df['Age'] <= 21]
old = df[df['Age'] > 21]

avg_young = young['Survived'].mean() * 100
avg_old = old['Survived'].mean() * 100

labely = (avg_old, avg_young)

labelx = 'young', 'old'

plt.bar(labelx, labely)
plt.xlabel('Age Group')
plt.ylabel('Percentage')
plt.title('Mortality Rate for Young and Old Survivors')
plt.show()

````
